```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

sns.set_style("white")
# best f1-score by different anomaly rate in hard identification approach
df = pd.DataFrame({'0.00': [0.1495, 0.0512, 0.1987],
                   '0.01': [0.2563, 0.2724, 0.5317],
                   '0.05': [0.4711, 0.3446, 0.6407],
                   '0.10': [0.5091, 0.3855, 0.6757],
                   '0.15': [0.5259, 0.3816, 0.6832],
                   'dataset': ['MSL', 'SMAP', 'ETL']})
df_long = df.melt(id_vars='dataset', var_name='measurement', value_name='value')

# draw linechart by seaborn
plt.figure(figsize=(7.5, 4.5))
fontsize = 15
sns.lineplot(data=df_long, x='measurement', y='value', hue='dataset', marker='o')
plt.xlabel(r'$\gamma$', fontsize=fontsize)
plt.ylabel('F1-Score', fontsize=fontsize)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(df.dataset,loc='upper left', fontsize=fontsize)
plt.show()
```
